```python
import pandas as pd

# Load dataset
df = pd.read_excel("Ecommerce_Sales_Analysis.xlsx")

# 1. Check basic structure
print(df.info())
print(df.head())

# 2. Check for missing values
print(df.isnull().sum())

# 3. Convert 'Order Date' to datetime
df["Order Date"] = pd.to_datetime(df["Order Date"])

# 4. Create new columns: Year-Month and Month
df["Year-Month"] = df["Order Date"].dt.to_period("M").astype(str)
df["Month"] = df["Order Date"].dt.month_name()
df["Year"] = df["Order Date"].dt.year

# 5. Save cleaned data
df.to_csv("Cleaned_Ecommerce_Data.csv", index=False)
```

```python
#STEP 4: Data Analysis 
```

```python
# Total Revenue
total_revenue = df["Total Price"].sum()

# Revenue by Category
category_sales = df.groupby("Product Category")["Total Price"].sum().sort_values(ascending=False)

# Monthly Sales
monthly_sales = df.groupby("Year-Month")["Total Price"].sum()

# Sales by Region
region_sales = df.groupby("Region")["Total Price"].sum()

# Top 5 Customers
top_customers = df.groupby("Customer ID")["Total Price"].sum().sort_values(ascending=False).head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar Chart: Category Sales
category_sales.plot(kind='bar', title='Revenue by Product Category', color='skyblue')
plt.ylabel('Revenue')
plt.xlabel('Product Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Line Chart: Monthly Sales Trend
monthly_sales.plot(kind='line', marker='o', title='Monthly Sales Trend')
plt.ylabel('Revenue')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Pie Chart: Regional Sales
region_sales.plot(kind='pie', autopct='%1.1f%%', title='Sales by Region')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
